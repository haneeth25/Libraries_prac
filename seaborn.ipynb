```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
import pandas as pd

# Create a dictionary with the data
data = {
    'ARG': [5.0, 12.0, 9.0, 9.0, 8.0],
    'BRA': [8.0, 1.0, 1.0, 4.0, 3.0],
    'ESP': [13.0, 14.0, 7.0, 7.0, 5.0],
    'FRA': [12.0, 7.0, 14.0, 15.0, 15.0],
    'GER': [1.0, 5.0, 4.0, 3.0, 1.0],
    'ITA': [2.0, 2.0, 3.0, 1.0, 2.0]
}

# Create a list of dates for the index
dates = pd.to_datetime(['1993-08-08', '1993-09-23', '1993-10-22', '1993-11-19', '1993-12-23'])

# Create the DataFrame
df = pd.DataFrame(data, index=dates)
df
```

```
df.head()
```

#We can plot all columns at a time using seaborn


```
plt.figure(figsize=(16,6))
sns.lineplot(df)
```

```
plt.figure(figsize=(16,6))
sns.lineplot(df)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```
plt.figure(figsize=(16,6))
sns.lineplot(df['ARG'],label = 'ARG')
sns.lineplot(df['BRA'],label = 'BRA')
plt.show()
```
