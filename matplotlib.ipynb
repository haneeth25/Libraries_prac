# piecharts

```
lang = ['c++','c#','python','java','go']
votes = [10,30,20,40,10]
plt.pie(votes,labels = lang)
plt.show()
```

```
explodes = [0.1,0,0,0.1,0]
```

```
plt.pie(votes,labels = lang,explode=explodes)
plt.show()
```

```
plt.pie(votes,labels = lang,explode=explodes,autopct = '%.2f%%',pctdistance=0.8)
plt.show()
```

```
plt.pie(votes,labels = lang,explode=explodes,autopct = '%.2f%%',pctdistance=0.8,startangle=90)
plt.show()
```

#Plot Customization

```
years = [2014+x for x in range(8)]
years
```

```
income = [55,56,62,61,72,72,73,75]
income_ticks = list(range(50,81,2))
```

```
plt.plot(years,income)
#title
plt.title('income of jhon (in usd)',fontsize = 30)
#xlabel
plt.xlabel('years')
#ylabel
plt.ylabel('yearly income(in usd)')
#to make y_axis into 55.0K usd instead of just showing 55.0
plt.yticks(income_ticks,[f"{x}k usd" for x in income_ticks])
#Similary we can do for x axis
plt.xticks(years,[f"{x}yy" for x in years],rotation = 45)
plt.show()
```

#Legends and multiple plots

```
import matplotlib.pyplot as plt
```

```
stock_a = [100,102,99,101,100,102]
stock_b = [90,95,102,104,105,103,109]
stock_c = [110,115,100,105,100,98,95]
plt.plot(stock_a,label='cmp1')
plt.plot(stock_b,label = 'cmp2')
plt.plot(stock_c,label = 'cmp3')
#plt.legend()
plt.legend(loc = 'lower center') #lower right
plt.show()
```

```
x1,y1 = np.random.random(100),np.random.random(100)
x2,y2 = np.arange(100),np.random.random(100)
```

```
plt.figure(1)
plt.scatter(x1,y1)
plt.figure(2)
plt.plot(x2,y2)
plt.show()
```

```
x = np.arange(100)
fig,axs = plt.subplots(2,2)
axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title('sine wave')

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title('cosine wave')

axs[1,0].plot(x,np.tan(x))
axs[1,0].set_title('tan wave')

axs[1,1].plot(x,np.log(x))
axs[1,1].set_title('log wave')
axs[1,1].set_xlabel('x')
axs[1,1].set_ylabel('y')
fig.suptitle('wave plots')
plt.show()
```
